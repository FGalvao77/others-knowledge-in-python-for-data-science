**Guia sobre wordcloud**

``Nuvem de Palavras [ Word Cloud ]``

Nuvem de palavras (word cloud) é um gráfico digital que mostra
o grau de frequência das palavras em um texto. Quanto mais a
palavra é utilizada, mais chamativa é a representação dessa
palavra no gráfico.

**Vamos importar as bibliotecas externas que precisamos**

```python
# importando as bibliotecas
import numpy as np                # para recursos matemáticos
import pandas as pd               # para modelagem de dados
import matplotlib.pyplot as plt   # para plotahgem de gráficos
from PIL import Image             # para carregar imagens
from wordcloud import WordCloud   # para gerar anuvem de palavras
```

**Definindo as palavras**

```python
# frases que vamos utilizar no exemplo
frase = '''
Lorem Ipsum é simplesmente uma simulação de texto da indústria tipográfica e
de impressos, e vem sendo utilizado desde o século XVI, quando um impressor
desconhecido pegou uma bandeja de tipos e os embaralhou para fazer um livro
de modelos de tipos. Lorem Ipsum sobreviveu não só a cinco séculos, como
também ao salto para a editoração eletrônica, permanecendo essencialmente
inalterado. Se popularizou na década de 60, quando a Letraset lançou
decalques contendo passagens de Lorem Ipsum, e mais recentemente quando
passou a ser integrado a softwares de editoração eletrônica como Aldus
PageMaker. 

Ao contrário da crença popular, Lorem Ipsum não é simplesmente
um texto aleatório. Tem raízes em uma peça da literatura clássica latina de 45
aC, com mais de 2.000 anos. Richard McClintock, um professor de latim no
Hampden-Sydney College, na Virgínia, pesquisou uma das palavras latinas mais
obscuras, consectetur, de uma passagem de Lorem Ipsum e, examinando as
citações da palavra na literatura clássica, descobriu a fonte indubitável. Lorem
Ipsum vem das seções 1.10.32 e 1.10.33 de "de Finibus Bonorum et Malorum"
(Os Extremos do Bem e do Mal) de Cícero, escrito em 45 aC. 

Este livro é um tratado sobre a teoria da ética, muito popular durante o Renascimento. A
primeira linha de Lorem Ipsum, "Lorem ipsum dolor sit amet ..", vem de uma
linha na seção 1.10.32.
'''
```

**Vamos definir a função da WordCloud**

```python
# gerando uma wordcloud (nuvem de palavras)
nuvem_palavras = WordCloud(
    # atribuindo uma cor de fundo 
    background_color = 'black',
    # definindo a largura
    width = 1600,
    # definindo a altura
    height = 800
).generate(frase) # atribuindo as palavras
```

**Criando um gráfico e atribuir as palavras**

```python
# instanciando o gráfico

# definindo o tamanho
fig, ax = plt.subplots(figsize = (10,7))
# atribuindo a nuvem de palavras em uma gráfico
# o comando "imshow" exibi os dados como uma iamgem ao invés de um gráfico
ax.imshow(nuvem_palavras, interpolation = 'bilinear')
# retirando os valores do gráfico
ax.set_axis_off()
```

**Vamos criar a Nuvem com uma imagem de fundo**
  - para isso, iremos importar uma imagem usando o Pillow (PIL)

```python
# carregar a iamgem
imagem_piramide = Image.open('piramide_1.jpg')
imagem_piramide
```

**Extraindo a intensidade de Pixels da imagem**

```python
# extraindo a itensidade de pixels
intensidade_pixels = np.array(imagem_piramide)
intensidade_pixels
```

>> Essa função do numpy irá extrair a intensidade de pixels da imagem.
Assim será gerado um array com os parâmetros da dimensão da imagem.

O wordcloud irá gerar as palavras em torno dessa intensidade de pixels.

**Vamos gerar a nuvem de palavras com o formato da pirâmide**

```python
# gerando a nuvem de palavras
nova_nuvem_palavras = WordCloud(
    # definindo a cor de fundo da imagem
    background_color = 'black',
    # definindo a largura
    width = 1500,
    # definindo a altura
    height = 1500,
    # definindo o número máximo de palavras
    max_words = 3000,
    # definindo a intensidade dos pixels
    mask = intensidade_pixels,
    # definindo a maior fonte
    max_font_size = 400,
    # definindo a menor fonte
    min_font_size = 1
).generate(frase)

# instanciando as configurações do gráfico
fig, ax = plt.subplots(figsize = (10,7))
# atribuindo a nuvem de palavras em um gráfico
ax.imshow(nova_nuvem_palavras, interpolation = 'bilinear')
# retirando os valores do gráfico
ax.set_axis_off()
```
