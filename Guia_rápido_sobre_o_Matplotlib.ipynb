**O que é o Matplotlib?**

``Matplotlib`` é uma biblioteca de software para criação de gráficos e visualizações de dados em geral, feita para e da linguagem de programação Python e sua extensão de matemática NumPy.

Originalmente criada pelo biólogo e neurocientista americano John D. Hunter, a biblioteca hoje possui uma comunidade ativa atuando em seu desenvolvimento e é distibuída sob uma licença BSD. O programador Michael Droetboom foi nomeado o líder do projeto um pouco antes da morte do criador John Hunter em agosto de 2012, e logo o cientista Thomas Caswell se juntou a ele.

- fonte: https://pt.wikipedia.org/wiki/Matplotlib


Podemos fazer diversos tipos de gráficos:
- Gráfico de linhas
- Gráfico de Barras

**Vamos ver suas funcionalidades em prática!**

```python
# importando as bibliotecas necessárias
import pandas as pd               # para modelagem de dados
import numpy as np                # para recursos matemáticos
import matplotlib.pyplot as plt   # para gerar recursos gráficos

%matplotlib inline
```

**Agora vamos gerar alguns números usando o ``numpy``**

```python
# gerando o números aleatórios
x = np.linspace(0, 10, 10)
y = x * x
```

Nesse exemplo estamos gerando números sequenciais usando a função ``linspace``.
- 1º Passamos o parâmetros para gerar valores do 0 a 10.
- 2º Passamos o parâmetro para gerar 10 valores.

Depois criamos outra variáveis e multiplicamos os valores. Assim teremos 2 eixos para gerar uma visão exponencial. Gráfico abaixo.

**Vamos gerar um gráfico de linhas, usando os recursos do ``Matplotlib``**

```python
# gráfico de linha com dois eixos 
# configurando os seus parâmetros
plt.figure(
    figsize=(14,9),   # tamanho do gráfico
    facecolor='lightblue'  # cor de fundo
)

# instanciando os eixos
ax = plt.axes()
ax.set_facecolor('white')   # cor de fundo do gráfico

# instanciando o gráfico
plt.plot(
    x, y,                       # dados do gráfico
    color='magenta',            # cor da linha
    linewidth=2,                # espessura da linha
    markersize=5,               # tamanho do marcador
    marker='o',                 # tipo do marcador
    linestyle='dashed',         # tipo da linha
    markeredgecolor='black'     # cor da linha do marcador
)

# configurando o label do eixo x
plt.xlabel(
    'eixo x',                   # nome do label
    fontweight='bold',          # estilo 
    fontsize='large',           # tamanho da fonte
    fontfamily='fantasy',       # tipo da fonte
    color='blue'                # cor da fonte
)

# configurando o label do eixo y
# iremos aplicar os mesmos parâmetros do label x
plt.ylabel('eixo y',             # nome do label
           fontweight='bold',    # estilo
           fontsize='large',     # tamanho da fonte
           fontfamily='fantasy', # tipo da fonte
           color='blue'          # cor da fonte
)

# definindo um título
plt.title(
    'Grafico de linha simples',   # título de gráfico
    fontweight='bold',            # estilo da fonte
    fontsize='xx-large',          # tamanho da fonte
    fontfamily='fantasy',         # tipo da fonte
    color='blue'                  # cor da fonte
)

# definindo as legendas
plt.legend(
    'Números aleatórios',                 # nome da legenda
    loc='upper center',                   # posição da legenda
    bbox_to_anchor=(1.0, 0.0, 0.0, 0.2),  # posição da legenda na figura
    fontsize='large',                     # tamanho do texto
    shadow=True,                          # sombra da legenda
    borderpad=1                           # borda da legenda
);
```

**Existe muitas opções para ser ajustada em um gráfico. Aqui exploramos algumas``mais usadas`` no dia a dia.**

Para conhecer outras configurações aconselho consultar a documentação oficial!

**Podemos criar uma grade com vários gráficos**

```python
# configurando os seus parâmetros
plt.figure(
    figsize=(14,9),   # tamanho do gráfico
    facecolor='lightgreen'  # cor de fundo
)

# definindo a dimensão da grade
linhas = 2
colunas = 2

# gráfico na posição 1 da grade
plt.subplot(linhas, colunas, 1)    # 1 é a posição na grade
plt.plot(x, 'r--')
plt.title('Gráfico 1')

# gráfico na posição 2 da grade
plt.subplot(linhas, colunas, 2)    # 2 é a posição na grade
plt.plot(y, 'g-')
plt.title('Gráfico 2')

# gráfico na posição 3 da grade
plt.subplot(linhas, colunas, 3)   # 3 é a posição na grade
plt.plot(y, 'y-')
plt.title('Gráfico 3')

# gráfico na posição 4 da grade
plt.subplot(linhas, colunas, 4)   # 4 é a posição na grade
plt.plot(x, 'b--')
plt.plot(y, 'p--')
plt.title('Gráfico 4');
```

Podemos gerar uma grade com diversos gráficos de vários tipos.

**Podemos gerar gráficos de diferentes tamanhos**

``Vamos criar um mini gráfico dentro de outro gráfico maior.``

```python
# configurando os seus parâmetros
plt.figure(
    figsize=(14,9),   # tamanho do gráfico
    facecolor='gray'  # cor de fundo
)

# instanciando a figura
fig = plt.figure()

# definindo o tamanho do mini gráfico 
eixo_1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])

# definindo o tamanho do gráfico maior
eixo_2 = fig.add_axes([0.2, 0.5, 0.3, 0.3])

# plotando os gráficos
eixo_1.plot(x, y, 'r--')
eixo_2.plot(x, y, 'y');
```

Podemos gerar diversos gráficos dentro de outros gráficos para detalhar alguma
informação ou expor outra informação.


**Podemos gerar outros formatos de gráficos**

``Vamos plotar um gráfico de barras``

```python
# criando a figura
fig = plt.figure()

# definindo o tamanho do gráfico
graf = fig.add_axes([0.0, 0.0, 0.7, 0.7])

# definindo os rótulos
rot = ['A','B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J']

# gerando o gráfico
graf.bar(rot, x);
```

``Agora vamos criar um gráfico de barras com mais eixos``

```python
# instanciando a figura
fig = plt.figure()

# definindo o tamanho
graf = fig.add_axes([0.0, 0.0, 0.7, 0.7])

# definindo os rótulos
rot = ['A','B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J']

# gerando o gráfico
graf.bar(rot, x, color='r')
graf.bar(rot, y * 0.1, color='b');
```

**Agora vamos conhecer outro tipo de gráfico**

``Boxplot``, conhecido popularmente de gráfico de caixa.

```python
# instanciando e plotando o boxplot
plt.boxplot(y);
```
