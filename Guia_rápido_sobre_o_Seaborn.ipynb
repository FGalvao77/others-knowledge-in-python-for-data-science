**O que é o Seaborn?**

``Seaborn`` é uma biblioteca de software para criação de gráficos e visualizações de dados em geral, feita para e a linguagem de programação ``Python``.

Podemos fazer diversos tipos de gráficos:
- Gráfico de linhas
- Gráfico de Barras
- Gráficos 3D
- E muitos outros.

**Vamos importar as bibliotecas que iremos pecisar**

```python
import numpy as np      # para recursos matemáticos
import pandas as pd     # para modelagem de dados
import seaborn as sns   # para análises gráfica

# definindo o tema "darkgrid" no seaborn
sns.set_theme(style='darkgrid')
```

**Vamos importar uma base de dados que já é incorporada no ``seaborn``**

```python
# instanciando a base de dados "tips"
base_dados = sns.load_dataset('tips')

# visualizando as 5 primeiras linhas
base_dados.head()
```

**Vamos renomear as colunas para deixar em ``português``**

```python
# traduzindo o nome das colunas
base_dados.rename(columns = {
    'total_bill' : 'total_conta',
    'tip' : 'gorjeta',
    'sex' : 'sexo',
    'smoker' : 'fumante',
    'day' : 'dia_semana',
    'time' : 'periodo',
    'size' : 'quantidade_pessoas_mesa'
}, inplace = True)
```

```python
# verificando se as colunas foram renomeadas
base_dados.columns
```

**Vamos utilizar o gráfico ``relplot``**

```python
# definindo os parêmetros do gráfico
sns.relplot(x = 'total_conta', 
            y = 'gorjeta',
            data = base_dados);
```

```python
# gráfico "regplot"
# com a função "regplot" e traçado a linha de regressão
sns.regplot(x = 'total_conta', 
            y = 'gorjeta',
            data = base_dados);
```

**Vamos usar o mesmo gráfico relplot e passar o parâmetro ``hue`**

```python
# gráfico "relplot"
'''
iremos passar outro argumento como classe no gráfico,
para isso iremos utilizar a coluna "sexo"
'''
sns.relplot(x = 'total_conta',
            y = 'gorjeta',
            data = base_dados,
            hue = 'sexo');
```

**Utilizar um gráfico de linhas com 2 eixos**

```python
# gráfico de linhas com dois eixos
sns.relplot(x = 'total_conta',
            y = 'gorjeta',
            data = base_dados,
            kind = 'line');
```

**Utilizando o gráfico de linhas com ``dois eixos`` e passando o parâmetro ``hue``**

```python
# gráfico de linhas com dois eixos
#  e com argumento "hue"
sns.relplot(x = 'total_conta',
            y = 'gorjeta',
            data = base_dados,
            kind = 'line',
            hue = 'fumante');
```

**Utilizando um gráfico de barras apenas com um eixo**

```python
sns.histplot(data = base_dados,
             x = 'total_conta');
```

**Utilizando o gráfico de barras apenas com um eixo e passando o parâmetro
``hue``**

```python
sns.histplot(data = base_dados, 
             x = 'total_conta',
             hue = 'fumante');
```

**Utilizando o gráfico ``barplot``**

```python
sns.barplot(data = base_dados,
            x = 'sexo',
            y = 'total_conta',
            hue = 'fumante');
```

**Utilizando o ``pairplot`` para gerar uma análise com todo o conjunto de dados**

```python
sns.pairplot(base_dados);
```

>> O ``Pairplot`` é um comando que gera diversos gráficos da base de dados,
desde de que sejam numéricos.

Caso tenha muitos dados, pode demorar bastante tempo para gerar as
análises gráficas.

**Utilizando o ``pairplot`` para gerar uma análise com todo o conjunto de dados e passando o parâmetro ``hue``**

```python
sns.pairplot(base_dados, hue = 'sexo');
```

>> O ``Pairplot`` também há uma serie de comandos para serem passados
como parâmetro.

Nesse exemplo passamos o parâmetro ``hue``.

**Utilizando o ``pairplot`` e passando o parâmetro ``kind``**

```python
sns.pairplot(base_dados, kind = 'kde');
```

**Utilizando o ``boxplot``**

```python
sns.boxplot(data = base_dados,
            x = 'dia_semana',
            y = 'total_conta');
```

**Utilizando o ``boxplot`` e passando o parâmetro ``hue``**

```python
sns.boxplot(data = base_dados,
            x = 'dia_semana',
            y = 'total_conta',
            hue = 'sexo');
```

```python
sns.boxplot(data = base_dados,
            x = 'dia_semana',
            y = 'total_conta',
            hue = 'fumante');
```

**Utilizando o ``violinplot``**

```python
sns.violinplot(data = base_dados,
               x = 'dia_semana',
               y = 'total_conta');
```

**Utilizando o ``swarmplot``**

```python
sns.swarmplot(data = base_dados,
              x = 'total_conta',
              y = 'sexo');
```

**Utilizando o ``joinplot``**

```python
sns.jointplot(data = base_dados, 
              x = 'total_conta', 
              y = 'gorjeta');
```
